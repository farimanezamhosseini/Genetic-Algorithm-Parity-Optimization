```python
import os 
os.system('python3 -m pip install numpy')
import numpy as np


#### Parameters 
n = int(input())
POPULATION_SIZE = 50
seed_range = np.random.randint(1001,size = POPULATION_SIZE)
arr_of_ints = np.random.randint(256,size = n)
arr_of_bools = np.random.randint(2,size = n)
# pc = 0.9
pm = 0.005
EPOCH = 500

#### odd or even
def odd_even(number):
    if np.mod(number,2) == 0:
        return 0
    else:
        return 1

#### Initial the population 
def init_population(arr_int,arr_bool,r):
    population = []
    for seed in r:
        entity = []
        np.random.seed(seed)
        entity.append(np.random.permutation(arr_int))
        entity.append(np.random.permutation(arr_bool))
        entity.append(seed)
        entity.append(None)
        population.append(entity)
    return population

#### Cross-over
def single_point_cross_over(population:list,n):
    population = list(population)
    # np.random.seed()
    children = []
    for i in range(0,len(population),2):
        child1 = []
        child2 = []
        rand = np.random.randint(1,n-1)
        child1_seed = population[i][2]
        child1_val = None
        child1_ints = list(population[i][0][:rand])
        child1_bools = list(population[i][1][:rand]) + list(population[i+1][1][rand:])
        for elem in population[i+1][0]:
            if elem not in child1_ints:
                child1_ints.append(elem)
        child1.append(child1_ints)
        child1.append(child1_bools)
        child1.append(child1_seed)
        child1.append(child1_val)
        child2_seed = population[i][2]
        child2_val = None
        child2_ints = list(population[i+1][0][rand:])
        child2_bools = list(population[i+1][1][rand:])+ list(population[i][1][:rand])
        for elem in population[i][0]:
            if elem not in child2_ints:
                child2_ints.append(elem)
        child2.append(child2_ints)
        child2.append(child2_bools)
        child2.append(child2_seed)
        child2.append(child2_val)
        children.append(child1)
        children.append(child2)
    return children

#### Mutaion
def mutation(population,pm):
    np.random.seed()
    for entity in population:
        for i in range(len(entity[1])):
            rand_float = np.random.random()
            if pm > rand_float:
                if entity[1][i] == 0:
                    entity[1][i] = 1
                else:
                    entity[1][i] = 0
    return population
### Fitness Evaluation function         
def fitness(population):
    for item in population:
        val = 0
        for i in range(len(item[0])):
            if odd_even(item[0][i]) == item[1][i]:
                val+=1
            else:
                pass
        item[-1] = val
    return population
```

```python
population = init_population(arr_of_ints,arr_of_bools,seed_range)
population = fitness(population)
population = sorted(population,key= lambda x : x[-1])[::-1]
if population[0][-1] == n:
    print(f"Best seed is found in initial population : the seed is = {population[0][-2]}")
else:
    for e in range(EPOCH):
        children = single_point_cross_over(population,n)
        # print(children)
        children = fitness(children)
        population = population + children
        population = sorted(population,key= lambda x : x[-1])[::-1]
        population = population[:POPULATION_SIZE]
        population = mutation(population,pm)
        population = fitness(population)
        population = sorted(population,key= lambda x : x[-1])[::-1]
        if population[0][-1] == n:
            print(f"Best seed is found in {e+1}th iteration : the seed is = {population[0][-2]} and the fitness value if {population[0][-1]}")
            print(f"Array of Integers : {population[0][0]}")
            print(f"Array of bools : {population[0][1]}")
            break
        else:
            print(f"Best seed so far in {e+1}th itration: {population[0][-2]} and fitness value equals to {population[0][-1]}")
    else:
        print(f"No seed got found in {EPOCH} iterations!!!")
        
```
